```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv ('https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv')
df = pd.DataFrame (data)
df.head()
```

```python
#1)By default kde=True,hist=True
sns.distplot(df['MaxTemp'])
```

```python
#2) Histogram
sns.distplot(df['MaxTemp'],kde=False,rug=True)
```

```python
#3)Kernel Density Estimation 1D plot
sns.distplot(df['MaxTemp'],hist=False,rug=True)
```

```python
#4) Kernel Density Estimation 1d plot
sns.kdeplot(df['MaxTemp'],shade=True)
```

```python
#5)
sns.kdeplot(df['MaxTemp'],label='original')
sns.kdeplot(df['MaxTemp'],bw=0.5)
sns.kdeplot(df['MaxTemp'],bw=2,label='bw=2')
plt.legend()
```

```python
#6)Scatter
sns.jointplot(kind='scatter',y=df['MaxTemp'],x=df['MinTemp'])
```

```python
#7) Hexbin Plot
sns.jointplot(kind='hex',y=df['MaxTemp'],x=df['MinTemp'])
```

```python
#8) Kernel Density Estimation 2D plot
sns.jointplot(kind = 'kde',y=df['MaxTemp'],x=df['MinTemp'])
```

```python
#9) Pair plot
x = pd.DataFrame(df,columns=['MaxTemp','MinTemp','Cloud9am','Cloud3pm','RainTomorrow'])
sns.pairplot(x)
```

```python
#10)
sns.pairplot(x,hue='RainTomorrow')
```

```python
#11)Strip plot
sns.stripplot(x='Cloud9am',y='Cloud3pm',data=df)
```

```python
#12)
sns.stripplot(x='Cloud9am',y='Cloud3pm',data=df,jitter=True)
```

```python
#13)Swarm Plot
sns.swarmplot(x='Cloud9am',y='Cloud3pm',data=df)
```

```python
#14)Box Plot
sns.boxplot(x='Cloud9am',y='Cloud3pm',data=df)
```

```python
#15)
sns.boxplot(x='RainToday',y='Humidity9am',data=df,hue='WindDir3pm',dodge=True)
```

```python
#16)
sns.boxplot(x='RainToday',y='Humidity9am',data=df,hue='WindDir3pm',dodge=False)
```

```python
#17)
sns.barplot(x='Cloud9am',y='Cloud3pm',data=df)
```

```python
#18)
sns.barplot(x='RainToday',y='Rainfall',hue='WindDir9am',data=df)
```

```python
#19)Count Plot
sns.countplot(x='MaxTemp',data=df)
```

```python
#20)
sns.countplot(x='WindDir3pm',data=df)
```
